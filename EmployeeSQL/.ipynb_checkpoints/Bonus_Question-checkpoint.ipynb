```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Connect to database
engine = create_engine("postgres://postgres:pace2013@localhost:5432/sql-challenge")
connect = engine.connect()
```

```python
#Call the salaries table
salaries_df = pd.read_sql("SELECT * FROM salaries", connect)
salaries_df.head()
```

```python
#Call the titles table
titles_df = pd.read_sql("SELECT * FROM titles", connect)
titles_df.head()
```

```python
#Join the tables (PK --> emp_no)
join_df = pd.merge(salaries_df, titles_df, on="emp_no", how="inner")
join_df.head()
```

```python
#Find the average by title
avg_df = join_df.groupby("title").mean()
avg_df.head()
```

```python
#Drop emp_no column, round to 2 decimal points, & reset index
clean_df = avg_df.drop(columns="emp_no")
final_df = clean_df.reset_index()
final_df.round(2)
# final_df.head()
```

```python
#Bar Chart
x_axis = final_df['title']
y_axis = final_df['salary']

plt.bar(x_axis, y_axis, align='center', alpha=0.5, color=['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'purple'])
ticks = [value for value in x_axis]
plt.xticks(ticks, x_axis, rotation='vertical')

plt.title("Avg Salary By Title")
plt.xlabel("Titles")
plt.ylabel("Salaries")

plt.savefig("../BonusPlot.png", bbox_inches='tight')

plt.show()
```
